# Introdução

Nesse experimento queremos estudar o movimento de um corpo rigido esférico sendo solto a partir do alto de uma canaleta e sendo liberado do contato ao final, de modo que realizará um movimento parabólico até entrar em contato com o solo.

Estudaremos a relação entre o alcance da esfera e a altura que a esfera foi solta. espera-se que o movimento a partir do ponto onde a esfera é liberada do contato com a canaleta até entrar em contato com o chão, se comporte de maneira independente vertical e horizontalmente onde a componente vertical depende exclusivamente da aceleração da gravidade e a componente horizontal depende exclusivamente da velocidade alcançada pelo corpo durante a descida pela canaleta. 

Em nosso modelo consideraremos duas possibilidades. Uma na qual o nosso corpo se comporta similarmente à uma particula de modo que desliza pela canaleta e toda sua energia potencial relativa à variação de altura entre o topo da canaleta e o final da canaleta é convertida em energia cinética tangencial. E a outra onde a bolinha se comporta como um corpo rígido e rola enquanto desce pela canaleta. Com base nessas previsões, mediremos de qual modelo teórico nosso corpo mais se aproxima. 

Em nosso modelo consideramos que a particula parte do repouso, portanto sua energia mecânica inicial é proporcional à sua altura com relação ao solo e sua energia cinética no instante que perde contato com a canaleta é proporcional à variação de altura entre o momento inicial e o momento de perda de contato, de modo que a energia cinética esperada para o corpo é 
$$
    E_{ctan} = E_{pi} = mgh \space Para \space o\space modelo\space sem\space rotação.\space \newline
    E_{crot} + E_{ctan} = E_{pi} = mgh \space Para \space o\space modelo\space com\space rotação.\space \newline
$$
Onde m é a massa do corpo, g é a aceleração da gravidade e h é a diferença de altura entre o ponto mais baixo e o ponto mais alto da canaleta.

Também sabemos que $E_{crot} = \frac{I\omega^2}{2}$ e $E_{ctan} = \frac{mv^2}{2}$ Logo,

$$
    \frac{mv^2}{2} = mgh \space Para \space o \space modelo \space sem\space rotação. \space \newline
    \frac{I\omega^2}{2} + \frac{mv^2}{2} = mgh \space Para \space o \space modelo \space com \space rotação.\space \newline
$$
onde v é a velocidade linear do objeto, $\omega$ é a velocidade angular da rotação do objeto e I é o momento de inércia do objeto 

Para o modelo sem rolamento, podemos descrever o alcance em função da velocidade ao sair da canaleta da seguinte maneira:


$$
    H = -\frac{g\cdot t^2}{2}
    \Rightarrow t = \sqrt{\frac{2H}{g}} \newline
    A = v \cdot t \Rightarrow
    A = v \cdot \sqrt{\frac{2H}{g}}
$$
Onde H é a distância entre o ponto mais baixo da canaleta, t é a variação de tempo entre o instante que a esfera perde contato com a canaleta e o instante onde a esfera entra em contato com o chão e A é o alcance do corpo.

E podemos colocar a velocidade ao perder o contato com a canaleta em função altura a partir da relação de conservação de energia mecânica da seguinte maneira:

$$
    \frac{v^2}{2} = gh \newline
    v = \sqrt{2gh}
$$


Logo

$$
    A = \sqrt{2gh} \cdot \sqrt{\frac{2H}{g}} \newline
    A = \sqrt{ 4 \cdot h \cdot H} \newline
    A^2 = 4 \cdot h \cdot H
$$


Já para o modelo onde a esfera rotaciona, teremos que considerar que parte da energia será convertida em energia cinética rotacional de modo que o alcance será afetado por tal. Sabendo que a canaleta não é perfeitamente plana e possui duas hastes em suas laterais, o raio de rotação da esfera será menor do que o raio da esfera em si, podendo ser descrito por:

$$
    R^2 = r^2 + d^2 \newline
    \Rightarrow r^2 = R^2 - d^2
$$

Onde R é o raio da esfera. r é o raio de rotação da esfera e d é metade da distância entre uma haste e outra da canaleta. (O esquema tanto da canaleta quanto da medição das dimensões da esfera estão disponíeis no método Experimental.)

Sabendo que a velocidade angular é diretamente proporcional à velocidade tangencial e ao raio de rotação, podemos descrever a velocidade angular da rotação da esfera da seguinte maneira: 

$$
    \omega = \frac{v}{r}
$$

Com base nisso afirmamos que: 

$$
    E_{crot} = \frac{I}{2} \cdot \frac{v^2}{r^2}
$$
