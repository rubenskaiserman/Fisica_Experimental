### Objetivo

Neste experimento, pretende-se medir experimentalmente a conservação do momento linear e a conservação da energia cinética ao análisar colisões elásticas e inelásticas.

Nota-se que é esperado observar a conservação da energia cinética e do momento linear na colisão elástica enquanto que esperamos que apenas o momento linear se conserve na colisão inelástica.

Com isso pretendemos verificar se o experimento se comporta exatamente como o previsto em um sistema ideal ou se há discrepâncias, e se, quais são as causas mais prováveis.

### Descrição do sistema

Colisão é definida como o momento em que dois corpos estão em contato, ou seja, se falamos de particulas, elas ocupam aproximadamente a mesma posição no espaço.

Dito isso, como experimentalmente estamos lidando com corpos extensos e estamos aproximando suas posições para a posição de seus centros de massa. Os objetos não ocuparão o mesmo espaço em nossa análise, como observaremos posteriormente. No caso da colisão inelástica esse resultado é ainda mais visível pois após colidirem, os objetos irão seguir mantendo uma distância fixa um do outro.

Nesse experimento poderemos observar que quando dois objetos colidem o momento linear do sistema como um todo se mantém, mas o momento linear de cada corpo pode variar por conta da colisão, onde um corpo transfere parte ou todo seu momento para o outro.

### Casos de estudo 
Em nossa análise, observaremos dois casos, em ambos existem dois corpos no sistema. Um corpo B em repouso e um corpo A em movimento retilíneo uniforme em direção a B.

Tanto no caso de colisão elástica quanto no caso de colisão inelástica, o sistema inicia com todo o momento linear e toda a energia cinética consentrada em A. Ou seja o momento linear e a energia cinética do sistema são descritos respectivamente por 
$$
    \vec{p} = m_a \cdot \vec{v_{ai}} \newline
    E_c = \frac{1}{2} \cdot m_{a}v_{ai}^{2}
$$

Note que embora representado como vetor, dado que o movimento do sistema em estudo é em apenas uma dimensão, é possível considerar o momento linear como um escalar dado que ele atua em apenas uma direção.

$$
    p = m_a \cdot v_{ai} \newline
    E_c = \frac{1}{2} \cdot m_{a}v_{ai}^{2}
$$

Observaremos que no caso de colisão elástica, ambas as grandezas se conservam, portanto,

$$
    p = m_a \cdot v_{ai} = m_a \cdot v_{af} + m_b \cdot v_{bf} \newline
    E_c = \frac{1}{2} \cdot m_{a}v_{ai}^{2} = \frac{1}{2} \cdot m_{a}v_{af}^{2} + \frac{1}{2} \cdot m_{b}v_{bf}^{2} 
$$

Já ná colisão inelástica observaremos que a energia cinética não se conserva, e poderemos medir a energia perdida na colisão através da seguinte equação

$$
    \Delta E_c = \frac{1}{2} \cdot m_{a}v_{ai}^{2} - (\frac{1}{2} \cdot m_{a}v_{af}^{2} + \frac{1}{2} \cdot m_{b}v_{bf}^{2})
$$
