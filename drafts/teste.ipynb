```python
from random import randint
import matplotlib.pyplot as plt 

results = []
for i in range(300):
    heads = 0
    for i in range(30):
        if randint(0,1) == 1:
            heads += 1
    results.append(heads)


results_counter = []
for i in range(max(results)*2+1):
    results_counter.append(results.count(i))

plt.bar(range(len(results)*2+1), results_counter, color='blue', align='center')
plt.show()
```
